# **Crimes contra o PatrimÃ´nio**

Esta seÃ§Ã£o serÃ¡ dedicada a anÃ¡lise dos crimes contra o patrimÃ´nio, ou seja, crimes de roubo, furto e recuperaÃ§Ã£o de veÃ­culos.

Os crimes contra o patrimÃ´nio sÃ£o aqueles que afetam os bens da vÃ­tima, sendo que a lei penal protege justamente os objetos da pessoa lesada. Entretanto, veremos a mÃ©dia com que esses crimes acontecem para analisar se realmente o bem Ã© protegido.

Assim, verificamos que a mÃ©dia de roubo de veÃ­culos Ã© de 2448.7 veÃ­culos por mÃªs.



```python
# ver a mÃ©dia da variÃ¡vel furto_veiculos
mean_furto_veiculos = df.furto_veiculos.mean()
print(mean_furto_veiculos)
```

Em relaÃ§Ã£o Ã  mÃ©dia de furto de veÃ­culos, observa-se que em comparaÃ§Ã£o com os roubos a mÃ©dia Ã© 1674.67, ou seja, furtos que Ã  princÃ­pio nÃ£o envolvem grave ameaÃ§a e violÃªncia a incidÃªncia Ã© baixa.

```python
mean_recuperacao_veiculos = df.recuperacao_veiculos.mean()
print(mean_recuperacao_veiculos)
```

JÃ¡ a mÃ©dia de veÃ­culos recuperados Ã© de 1772.23. 

Vamos entÃ£o analisar a porcentagem de carros recuperados em relaÃ§Ã£o aos carros roubados e os furtados, alÃ©m de encontrar os valores mÃ¡ximos e mÃ­nimos de homicÃ­dios dolosos.

```python
# ver a mÃ©dia da variÃ¡vel recuperacao_veiculos
mean_recuperacao_veiculos = df.recuperacao_veiculos.mean()

# total de veÃ­culos perdidos
total = mean_recuperacao_veiculos + mean_furto_veiculos
print('MÃ©dia de VeÃ­culos Perdidos por MÃªs: {:.2f}'.format(total))
print('VeÃ­culos Recuperados por MÃªs: {:.2f}'.format(mean_recuperacao_veiculos))

# porcentagem de veÃ­culos recuperados
pct = mean_recuperacao_veiculos/total*100

print('{:.2f}% dos veÃ­culos roubados ou furtados sÃ£o recuperados.'.format(pct))
```

Ao verificar a mÃ©dia de veÃ­culos em recuperaÃ§Ã£o sobre os roubados e furtados, temos que 3.446.91 veÃ­culos sÃ£o perdidos, ou seja, a polÃ­cia deixa de resgatar mais da metade dos veÃ­culos, 51.42%, tendo em vista a violÃªncia na cidade.
